### Points Table Scraping from livemint.com

```python
url="https://www.livemint.com/sports/cricket-news/ipl-points-table"
r= requests.get(url)
soup = BeautifulSoup(r.text,"lxml")

div1=soup.find("div",class_="customTblMain")

table1=div1.find("table")

headers1=["Pos","Logo","Teams"]
rows1=[]
# for header in table1.find_all("th"):
#     headers1.append(header.text.strip())
    
for row in table1.find_all('tr'):
    row_data = []
    for cell in row.find_all(["th","td"]):
        row_data.append(cell.text.strip())
    for cell in row.find_all("strong",class_="longNam"):
        cell=cell.text.strip().upper()
        row_data.append(cell)
    for cell in row.find_all("strong",class_="shortNam"):
        cell=cell.text.strip().upper()
        img=f"https://scores.iplt20.com/ipl/teamlogos/{cell}.png"
        row_data[1]=img
    rows1.append(row_data)
rows1=rows1[1:]

df1=pd.DataFrame(rows1,columns=headers1)

div2=soup.find("div",class_="scrollTabl")
table=div2.find("table")

headers2=[]
rows2=[]
for header in table.find_all("th"):
    headers2.append(header.text.strip())
    
for row in table.find_all('tr'):
    row_data = []
    for cell in row.find_all(["th","td"]):
        row_data.append(cell.text.strip())
    for cell in row.find_all("strong",class_="longNam"):
        cell=cell.text.strip()
        row_data[1]=cell
    for cell in row.find_all("strong",class_="shortNam"):
        cell=cell.text.strip().upper()
        row_data.append(cell)
    rows2.append(row_data)
rows2=rows2[1:]

df2=pd.DataFrame(rows2,columns=headers2)
df2=df2.iloc[:,:-1]

df = pd.concat([df1, df2], axis=1)

df
```

### Orange Cap List from livemint.com

### Making complete dataset

```python
import pandas as pd

orange=pd.read_csv("Orange_cap.csv")
purple=pd.read_csv("Purple_cap.csv")

df=pd.merge(orange,purple,on='Name',how="outer",suffixes=('_orange', '_purple'))

df["T_orange"]=df["T_orange"].fillna("")
df["T_purple"]=df["T_purple"].fillna("")
df["Player_orange"]=df["Player_orange"].fillna("")
df["Player_purple"]=df["Player_purple"].fillna("")

def combine_teams(row):
    if row['T_orange'] == '':
        return row['T_purple']
    elif row["T_purple"]=="":
        return row['T_orange']
    else:
        return row["T_orange"]
    
def combine_profiles(row):
    if row['Player_orange'] == '':
        return row['Player_purple']
    elif row["Player_purple"]=="":
        return row['Player_orange']
    else:
        return row["Player_orange"]
    
df['Team'] = df.apply(combine_teams, axis=1)
df['Player'] = df.apply(combine_profiles, axis=1)

df=df.drop(["T_orange","T_purple","ID_orange",'ID_purple',"Player_purple","Player_orange"],axis=1)

df=df.fillna(0)

columns={"R_orange":"Runs_scored","SR_orange":"Bat_Strike","Avg_orange":"Bat_Avg","W":"Wickets_Taken","Avg_purple":"Ball_Avg","R_purple":"Runs_conceeded","SR_purple":"Ball_Strike"}

df=df.rename(columns=columns)

df
```

### Scraping matches dataset from wikipedia

```python
url="https://en.wikipedia.org/wiki/2024_Indian_Premier_League"
r= requests.get(url)
soup = BeautifulSoup(r.text,"lxml")

table=soup.find("div",class_="mw-content-ltr mw-parser-output")

def has_style(tag, style):
    return tag.has_attr('style') and style in tag['style']

scores=[]
initial_a_elements = table.find_all("div")
for tables in initial_a_elements:
    t=tables.find_all("table")
    style_to_find = "float:left; width:53%; table-layout:fixed; text-align:center; border-style:none"
    elements_with_style = tables.find_all(lambda tag: has_style(tag, style_to_find))
    for x in elements_with_style:
        style="vertical-align:top"
        y=x.find_all(lambda tag:has_style(tag,style))
        i=0
        sc=[]
        for z in y:
            # st="width:47%; text-align:right; border-style:none"
            # j=z.find_all(lambda tag:has_style(tag,st))
            i+=1
            if i%2!=0:
                a=z.find_all("td")
                num=0
                for j in a:
                    num+=1
                    for k in j:
                        s=""
                        for l in k:
                            try:
                                if num%2!=0:
                                    s+=l.text.strip()+"|"
                            except:
                                pass
                        sc.append(s)
        scores.append(sc)
scores1=[]
for i in scores:
    scores1.append(i[0]+"||"+i[4])

initial_a_elements = table.find_all("div")
match_nos=[]
for tables in initial_a_elements:
    t=tables.find_all("table")
    style_to_find = "float:left; width:15%; table-layout:fixed; border-style:none"
    elements_with_style = tables.find_all(lambda tag: has_style(tag, style_to_find))
    for x in elements_with_style:
        style_to_find = "text-align:right"
        elements = x.find_all(lambda tag: has_style(tag, style_to_find))
        for y in elements:
            ma=y.find("b")
            # print(ma.text.strip())
            match_nos.append(ma.text.strip())
new_score=[]
for i in range(len(match_nos)):
    new_score.append([match_nos[i],scores1[i]])

ven=[]
initial_a_elements = table.find_all("div")
for tables in initial_a_elements:
    t=tables.find_all("table")
    style_to_find = "float:right; width:32%; overflow:hidden; border-style:none"
    elements_with_style = tables.find_all(lambda tag: has_style(tag, style_to_find))
    if elements_with_style!=[]:
        for e in elements_with_style:
            # l=[e.text.strip()]
            # for i in l:
            #     print(i[:i.index("Umpires:")])
            a_tags=e.find_all("a")
            ven.append(e.find("b").text.strip()+"|"+str(a_tags[0].text.strip())+"|"+str(a_tags[1].text.strip()))

new_scores=[]
for i in range(74):
    new_scores.append(new_score[i]+[ven[i]])
new_scores

for i in new_scores:
        i.extend([i[1].split("|||")[0],i[1].split("|||")[1]])
        try:
            i.extend([i[1].split("||||")[0],i[1].split("||||")[1]])
        except:
             pass
for i in new_scores:
    try:
        i.extend([i[3].split("||")[0],i[3].split("||")[1]])
        i.extend([i[4].split("||")[0],i[4].split("||")[1]])
    except:
        pass
    i.extend([i[2].split("|")[0],i[2].split("|")[1],i[2].split("|")[2]])

import pandas as pd

df=pd.DataFrame(new_scores,columns=["Match No","TT","SS","TT1","TT2","Team1","Score1","Team2","Score2","Summary","Stadium","City"])

df=df.drop(["TT","SS","TT1","TT2"],axis=1)

df
```
